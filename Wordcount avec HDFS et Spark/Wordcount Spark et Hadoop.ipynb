- Maintenant passons au programme qui va faire le wordcount sur le ficihier texte shakespear:

```python
# Accéder a votre fichier sur hadoop et stocker ces données dans une variable 
file = sc.textFile("hdfs:/input_data/shakespear.txt")
```

```python
# Accéder a votre fichier sur hadoop et stocker ces données dans une variable 
file = sc.textFile("hdfs:/input_data/shakespear.txt")
# Programme qui execute le word count sur notre fichier
#1. Décomposition de chaque ligne en mots
counts = file.flatMap(lambda line: line.split(" ")).map(lambda word: (word, 1)).reduceByKey(lambda v1, v2: v1 + v2)
#2. Chacun des mots est transformé en une clé-valeur 
    #.map(lambda word: (word, 1))
#3. Les valeurs associées à chaques clé sont sommées
    #.reduceByKey(lambda v1, v2: v1 + v2)
#4. Le résultat est récupéré
    #.collect()
```

```python
# Chaque paire (clé, valeur) est affichée
for (word, count) in counts.collect():
    print('{}:{} \n'.format(word, count))
```

```python
# On enregistre le resultat sous Hadoop HDFS
counts.saveAsTextFile("hdfs:/theoutput")
```
